```python
import pandas as pd

# Load dataset
df_hindi_english = pd.read_excel(r"C:\Users\yashw\Downloads\hindi_english_movies_dataset_70_records (1)2.xlsx")

# 1. Remove duplicates (same movie repeated with suffixes like (1), (2), (3))
df_hindi_english['Movie Name'] = df_hindi_english['Movie Name'].str.replace(r"\(\d+\)", "", regex=True)
df_hindi_english = df_hindi_english.drop_duplicates()

# 2. Round ratings to one decimal place
df_hindi_english['Rating'] = df_hindi_english['Rating'].round(1)

# 3. Standardize genres (optional: replace slashes with commas)
df_hindi_english['Genre'] = df_hindi_english['Genre'].str.replace("/", ", ")

# 4. Reset index
df_hindi_english = df_hindi_english.reset_index(drop=True)

# Save cleaned dataset
df_hindi_english.to_csv("Hindi_English_Movies_Cleaned.csv", index=False)

print(df_hindi_english.head(10))
```

```python
import pandas as pd

# Load Hindi-English dataset
hindi_english = pd.read_csv(r"C:\Users\yashw\Downloads\Hindi_English_Movies_Cleaned (1).csv")

# Load Telugu dataset
telugu = pd.read_csv(r"C:\Users\yashw\Downloads\Telugu_Movies_Cleaned (1).csv")

# Standardize column names for merging
telugu = telugu.rename(columns={
    "title": "Movie Name",
    "genre": "Genre",
    "director": "Director",
    "year": "Year",
    "runtime (min)": "Runtime (min)",
    "rating": "Rating"
})

# Add missing columns to Telugu dataset to align with Hindi-English dataset
for col in ["Language", "Lead Actor", "Lead Actress", "Review"]:
    telugu[col] = "N/A"

# Assign language explicitly
telugu["Language"] = "Telugu"

# Merge datasets (stacking them together)
merged_movies = pd.concat([hindi_english, telugu], ignore_index=True)

# Save merged dataset
merged_movies.to_csv("Merged_Movies.csv", index=False)

print("Merged dataset created successfully with shape:", merged_movies.shape)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Hindi-English dataset
hindi_english = pd.read_csv(r"C:\Users\yashw\Downloads\Hindi_English_Movies_Cleaned (1).csv")

# Load Telugu dataset
telugu = pd.read_csv(r"C:\Users\yashw\Downloads\Telugu_Movies_Cleaned (1).csv")

# Standardize Telugu column names
telugu = telugu.rename(columns={
    "title": "Movie Name",
    "genre": "Genre",
    "director": "Director",
    "year": "Year",
    "runtime (min)": "Runtime (min)",
    "rating": "Rating"
})

# Add missing columns to Telugu dataset
for col in ["Language", "Lead Actor", "Lead Actress", "Review"]:
    telugu[col] = "N/A"

# Assign language explicitly
telugu["Language"] = "Telugu"

# Merge datasets
merged_movies = pd.concat([hindi_english, telugu], ignore_index=True)

# Save merged dataset
merged_movies.to_csv("Merged_Movies.csv", index=False)

print("Merged dataset created successfully with shape:", merged_movies.shape)

# --- Visualization: Language-wise movie count ---
plt.figure(figsize=(8,5))
merged_movies["Language"].value_counts().plot(kind="bar", color=["orange","blue","green"])
plt.title("Movie Count by Language")
plt.xlabel("Language")
plt.ylabel("Number of Movies")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load datasets
telugu_df = pd.read_csv()
hindi_eng_df = pd.read_csv("Hindi_English_Movies_Cleaned.csv")

# Standardize Telugu dataset columns to match Hindi/English dataset
telugu_df = telugu_df.rename(columns={
    "title": "Movie Name",
    "genre": "Genre",
    "director": "Director",
    "year": "Year",
    "runtime (min)": "Runtime (min)",
    "rating": "Rating"
})

# Add missing columns to Telugu dataset (fill with NaN or placeholders)
telugu_df["Language"] = "Telugu"
telugu_df["Lead Actor"] = None
telugu_df["Lead Actress"] = None
telugu_df["Review"] = None

# Ensure Hindi/English dataset has consistent column names
hindi_eng_df = hindi_eng_df.rename(columns={
    "Movie Name": "Movie Name",
    "Language": "Language",
    "Year": "Year",
    "Genre": "Genre",
    "Rating": "Rating",
    "Runtime (min)": "Runtime (min)",
    "Director": "Director",
    "Lead Actor": "Lead Actor",
    "Lead Actress": "Lead Actress",
    "Review": "Review"
})

# Merge both datasets
combined_df = pd.concat([telugu_df, hindi_eng_df], ignore_index=True)

# Example analysis: average rating by language
avg_rating_by_lang = combined_df.groupby("Language")["Rating"].mean()

print("Combined Dataset Shape:", combined_df.shape)
print("\nAverage Rating by Language:\n", avg_rating_by_lang)

# Save merged dataset if needed
combined_df.to_csv("Integrated_Movies.csv", index=False)
```
